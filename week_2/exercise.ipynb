##imports

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Addition of image

```python
def addtion_image(username, img, added_image="profiles/landscape.jpg"):
    image1 = cv2.imread(img)
    image2 = cv2.imread(added_image)
    
    
    image1 = cv2.resize(image1, (500,400))
    image2 = cv2.resize(image2, (500,400))
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image1)
    
    addimage = cv2.addWeighted(image1, 0.5, image2, 0.6, 0)
    
    plt.subplot(1, 2, 2)
    plt.title("Addition of image")
    plt.imshow(addimage)
    plt.show()
    # cv2.imshow("Weighted Image", addimage)
    # if cv2.waitKey(0) & 0xff == 27:
    #     cv2.destroyAllWindows()
    
```

```python
addtion_image("zion", "profiles/lala.jpg")
```

### Brightness & Contrast

```python
def bright_contrast(username, img, bright_num, contrast_num):
    image = cv2.imread(img)
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    image2 = cv2.addWeighted(image, contrast_num, np.zeros(image.shape, image.dtype), 0, bright_num)
    
    # cv2.imwrite(f"processed/{username}_bright.jpg", image2)
    
    plt.subplot(1, 2, 2)
    plt.title("Brightness & Contrast")
    plt.imshow(image2)
    plt.show()
```

```python
bright_contrast("zion", "profiles/lala.jpg", 5, 1.5)
```

### Sharpening Images

```python
def sharpening(username, img):
    image = cv2.imread(img)
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    
    sharpened_image = cv2.filter2D(image, -1, kernel, kernel)
    
    cv2.imwrite(f"processed/{username}_sharpened.jpg", sharpened_image)
    
    plt.subplot(1, 2, 2)
    plt.title("Sharpening")
    plt.imshow(sharpened_image)
    plt.show()
```

```python
sharpening("zion", "profiles/lala.jpg")
```

### Remove noise

```python
def remove_noise(username, img):
    image = cv2.imread(img)
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    filtered_image = cv2.medianBlur(image, 15)
    
    
    cv2.imwrite(f"processed/{username}_thresh.jpg", filtered_image)
    
    plt.subplot(1, 2, 2)
    plt.title("Median Blur")
    plt.imshow(filtered_image)
    plt.show()
```

```python
remove_noise("zion", "profiles/lala.jpg")
```

### Image Scaling

```python
def img_scaling(username, img, x, y):
    image = cv2.imread(img)
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    resized_image = cv2.resize(image, None, fx=x,fy=y)
    
    cv2.imwrite(f"processed/{username}_resized.jpg", resized_image)
    
    plt.subplot(1, 2, 2)
    plt.title("Scaled")
    plt.imshow(resized_image)
    plt.show()
```

```python
img_scaling("zion", "profiles/lala.jpg", 2, 2)
```

### Inverse Transform

```python
def inverse_transform(username, img):
    image = cv2.imread(img)
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)   
    
    inverse_image = 255 - image
    
    cv2.imwrite(f"processed/{username}_inverse.jpg", inverse_image)
    
    plt.subplot(1, 2, 2)
    plt.title("Inverse color")
    plt.imshow(inverse_image)
    plt.show()
```

```python
inverse_transform("zion", "profiles/lala.jpg")
```

```python
students = {
    "chudi": {"matno": "001", "image": "chudi.jpg"},
    "eze": {"matno": "002", "image": "eze.jpg"},
    "george": {"matno": "003", "image": "george.jpg"},
    "kaobi": {"matno": "004", "image": "kaobi.jpg"},
    "lala": {"matno": "005", "image": "lala.jpg"},
    "nifemi": {"matno": "006", "image": "nifemi.jpg"},
    "ogb": {"matno": "007", "image": "ogb.jpg"},
    "oyinda": {"matno": "008", "image": "oyinda.jpg"},
    "sope": {"matno": "009", "image": "sope.jpg"},
    "tets": {"matno": "010", "image": "tets.jpg"}
}
```

```python
def perform_enhancements(username, image_path, enhancements):
    # Simulated function to perform enhancements
    print(f"Performing enhancements on {image_path}...")
    print(enhancements)
    for enhancement in enhancements:
        if enhancement == "Addition of image":
            addtion_image(username, image_path)
        elif enhancement == "Brightness & contrast":
            bright_contrast(username, image_path, 5, 1.5)
        elif enhancement == "Sharpening":
            sharpening(username, image_path)
        elif enhancement == "Removing noise":
            remove_noise(username, image_path)
        elif enhancement == "Image Scaling":
            img_scaling(username, image_path, 2, 2)
        elif enhancement == "Inverse Transform":
            inverse_transform(username, image_path)
        else:
            print(f"{enhancement} is not a valid enhancement")
            
            
```

```python
def login():
    while True:
        username = input("Enter your username: ")
        matno = input("Enter your matriculation number: ")
        if username in students and students[username]["matno"] == matno:
            return username
        else:
            print("Invalid username or matriculation number. Please try again.")
```

```python
def main():
    username = login()
    print(f"Welcome, {username}!")
    image_name = students[username]["image"]
    image_path = f"profiles/{image_name}"  # Replace "path_to_images" with the actual path to images

    print("Available enhancements:")
    print("1. Addition of image")
    print("2. Brightness & contrast")
    print("3. Sharpening")
    print("4. Removing noise")
    print("5. Image Scaling")
    print("6. Inverse Transform")

    selected_enhancements = []
    while True:
        enhancement_choice = input("Enter the number corresponding to the enhancement you want (or type 'done' to finish): ")
        if enhancement_choice.lower() == "done":
            break
        elif enhancement_choice.isdigit() and 1 <= int(enhancement_choice) <= 6:
            selected_enhancements.append(enhancement_choice)
        else:
            print("Invalid input. Please enter a number between 1 and 6 or 'done'.")

    # Convert enhancement choices to actual enhancement names
    enhancement_mapping = {
        "1": "Addition of image",
        "2": "Brightness & contrast",
        "3": "Sharpening",
        "4": "Removing noise",
        "5": "Image Scaling",
        "6": "Inverse Transform"
    }
    selected_enhancements = [enhancement_mapping[choice] for choice in selected_enhancements]

    # Perform enhancements
    perform_enhancements(username, image_path, selected_enhancements)
    print("Enhancements completed successfully!")
```

```python
main()
```
