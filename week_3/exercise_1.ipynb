### Imports

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

## Transformations

```python
def image_translation(img):
    image = cv.imread(img, 0)
    rows, cols = image.shape
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    M = np.float32([[1, 0, 100], [0, 1, 50]])
    trans1 = cv.warpAffine(image, M, (cols, rows))
    
    plt.subplot(1,2, 2)
    plt.title("Translated Image")
    plt.imshow(trans1)
    
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
def reflected_image(img):
    image = cv.imread(img, 0)
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    rows, cols = image.shape
    M = np.float32([[1, 0, 0], [0, -1, rows], [0, 0, 1]])
    reflect_img = cv.warpPerspective(image, M, (int(cols), int(rows)))
    
    plt.subplot(1,2, 2)
    plt.title("Reflected Image")
    plt.imshow(reflect_img)
    
    # cv.imwrite('relfection_out.jpg', reflect_img)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
def image_rotation(img):
    image = cv.imread(img, 0)
    rows, cols = image.shape
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    
    
    rot_img = cv.warpAffine(image, cv.getRotationMatrix2D((cols/2, rows/2), 30, 0.6), (cols, rows))
    
    plt.subplot(1,2, 2)
    plt.title("Rotated Image")
    plt.imshow(rot_img)
    
    # cv.imshow('img', rot_img)
    # cv.imwrite('rotation_out.jpg', rot_img)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
def image_cropping(img):
    image = cv.imread(img, 0)
    # rows, cols = image.shape
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    crop_img = image[450:700, 200:500]
    
    plt.subplot(1,2, 2)
    plt.title("Cropped Image")
    plt.imshow(crop_img)
    
    # cv.imshow('img', crop_img)
    # cv.imwrite('cropping_out.jpg', crop_img)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
def image_sheared_x(img):
    image = cv.imread(img, 0)
    rows, cols = image.shape
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
    shear_img = cv.warpPerspective(image, M, (int(cols*1.5), int(rows*1.5)))
    
    plt.subplot(1,2, 2)
    plt.title("Sheared Image")
    plt.imshow(shear_img)
    
   
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
def image_sheared_y(img):
    image = cv.imread(img, 0)
    rows, cols = image.shape
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    M = np.float32([[1, 0, 0], [0.5, 1, 0], [0, 0, 1]])
    shear_img = cv.warpPerspective(image, M, (int(cols*1.5), int(rows*1.5)))
    
    plt.subplot(1,2, 2)
    plt.title("Sheared Image Y-axis")
    plt.imshow(shear_img)
    
    # cv.imshow("img",shear_img)
    cv.waitKey(0)
    cv.destroyAllWindows()
    
    
```

```python
def image_blurring(img):
    image = cv.imread(img)
    
    cv.imshow("Original image", image)
    cv.waitKey(0)
    
    
    
    
    # Gaussian blur
    Gaussian = cv.GaussianBlur(image, (7,7), 0)
    cv.imshow("Gaussian Blurring", Gaussian)
    cv.waitKey(0)
    
    # Median blur
    median = cv.medianBlur(image, 5)
    cv.imshow("Median Blurring", median)
    cv.waitKey(0)
    
    # Bilateral filter
    bilateral = cv.bilateralFilter(image, 9, 75, 75)
    cv.imshow("Bilateral Blurring", bilateral)
    cv.waitKey(0)
    
    cv.destroyAllWindows()
```

```python
def login():
    while True:
        email = input("Enter your email: ")
        if email.endswith("@pau.edu.ng"):
            age = int(input("Enter your age: "))
            if age >= 18:
                return email
            else:
                print("You must be 18 years or older to access the collection.")
        else:
            print("Invalid email. Please use an email ending with @pau.edu.ng.")
```

```python
def perform_enhancements(img,transformations):
    # Simulated function to perform enhancements
    for transformation in transformations:
        if transformation == "Image Translation":
            image_translation(img)
        elif transformation == "Image Reflection":
            reflected_image(img)
        elif transformation == "Image Rotation":
            image_rotation(img)
        elif transformation == "Image Cropping":
            image_cropping(img)
        elif transformation == "Sheared Image along X-axis":
            image_sheared_x(img)
        elif transformation == "Sheared Image along Y-axis":
            image_sheared_y(img)
        elif transformation == "Image Blurring":
            image_blurring(img)
        else:
            print(f"{transformation} is not a valid transformation.")
```

```python
def main():
    transformation_mapping = {
        "1": "Image Translation",
        "2": "Image Reflection",
        "3": "Image Rotation",
        "4": "Image Cropping",
        "5": "Sheared Image along X-axis",
        "6": "Sheared Image along Y-axis",
        "7": "Image Blurring"
    }
    
    email = login()
    print(f"Welcome, {email}!")
    
    while True:
        category = select_art_category()
        if category:
            artwork, transformations = select_artwork_and_transformations(category)
            print(f"the artwork is {artwork}")
            selected_transformations = [transformation_mapping[choice] for choice in transformations]
            perform_enhancements(artwork, selected_transformations)
            break
        else:
            break
```
