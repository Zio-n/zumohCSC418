### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

### Image Translation

```python
def image_translation(img):
    image = cv.imread(img, 0)
    rows, cols = image.shape
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    M = np.float32([[1, 0, 100], [0, 1, 50]])
    trans1 = cv.warpAffine(image, M, (cols, rows))
    
    plt.subplot(1,2, 2)
    plt.title("Translated Image")
    plt.imshow(trans1)
    
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
image_translation("images/kaobi.jpg")
```

### Reflection

```python
def reflected_image(img):
    image = cv.imread(img, 0)
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    rows, cols = image.shape
    M = np.float32([[1, 0, 0], [0, -1, rows], [0, 0, 1]])
    reflect_img = cv.warpPerspective(image, M, (int(cols), int(rows)))
    
    plt.subplot(1,2, 2)
    plt.title("Reflected Image")
    plt.imshow(reflect_img)
    
    # cv.imwrite('relfection_out.jpg', reflect_img)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
reflected_image("images/kaobi.jpg")
```

### Image Rotation

```python
def image_rotation(img):
    image = cv.imread(img, 0)
    rows, cols = image.shape
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    
    
    rot_img = cv.warpAffine(image, cv.getRotationMatrix2D((cols/2, rows/2), 30, 0.6), (cols, rows))
    
    plt.subplot(1,2, 2)
    plt.title("Rotated Image")
    plt.imshow(rot_img)
    
    # cv.imshow('img', rot_img)
    # cv.imwrite('rotation_out.jpg', rot_img)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
image_rotation("images/kaobi.jpg")
```

### Image Cropping

```python
def image_cropping(img):
    image = cv.imread(img, 0)
    # rows, cols = image.shape
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    crop_img = image[450:700, 200:500]
    
    plt.subplot(1,2, 2)
    plt.title("Cropped Image")
    plt.imshow(crop_img)
    
    # cv.imshow('img', crop_img)
    # cv.imwrite('cropping_out.jpg', crop_img)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
image_cropping("images/kaobi.jpg")
```

### Sheared Image X-axis

```python
def image_sheared_x(img):
    image = cv.imread(img, 0)
    rows, cols = image.shape
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
    shear_img = cv.warpPerspective(image, M, (int(cols*1.5), int(rows*1.5)))
    
    plt.subplot(1,2, 2)
    plt.title("Sheared Image")
    plt.imshow(shear_img)
    
   
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
image_sheared_x("images/kaobi.jpg")
```

### Sheared Image Y-Axis

```python
def image_sheared_y(img):
    image = cv.imread(img, 0)
    rows, cols = image.shape
    
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    
    M = np.float32([[1, 0, 0], [0.5, 1, 0], [0, 0, 1]])
    shear_img = cv.warpPerspective(image, M, (int(cols*1.5), int(rows*1.5)))
    
    plt.subplot(1,2, 2)
    plt.title("Sheared Image Y-axis")
    plt.imshow(shear_img)
    
    # cv.imshow("img",shear_img)
    cv.waitKey(0)
    cv.destroyAllWindows()
    
    
```

```python
image_sheared_y("images/kaobi.jpg")
```

### Image Blurring

```python
def image_blurring(img):
    image = cv.imread(img)
    
    cv.imshow("Original image", image)
    cv.waitKey(0)
    
    
    
    
    # Gaussian blur
    Gaussian = cv.GaussianBlur(image, (7,7), 0)
    cv.imshow("Gaussian Blurring", Gaussian)
    cv.waitKey(0)
    
    # Median blur
    median = cv.medianBlur(image, 5)
    cv.imshow("Median Blurring", median)
    cv.waitKey(0)
    
    # Bilateral filter
    bilateral = cv.bilateralFilter(image, 9, 75, 75)
    cv.imshow("Bilateral Blurring", bilateral)
    cv.waitKey(0)
    
    cv.destroyAllWindows()
```

```python
image_blurring("images/kaobi.jpg")
```
